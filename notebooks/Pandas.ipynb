```python
import pandas as pd

# Opening file in pandas
df = pd.read_csv('drive/MyDrive/data/assets/penguins.csv')

df.dropna(inplace=True)

def get_memuza_of_mass(df) -> float:
  bodysum = 0
  for i in df.body_mass_g:
    try:
      if not float(i) > 0:
        continue
      bodysum += i
    except Exception as e:
      continue
  return bodysum / len(df.body_mass_g)



#for i in df["species"].unique():
  #print(f"Average of {i} is {get_memuza_of_mass(df[df['species'] == i])}")

females = df[df['sex'] == "female"]
print(f"Average of females is {get_memuza_of_mass(females)}")

```

```python
array = [4.8, 4.72, 4.88, 4.71, 4.91, 5, 5, 5, 5, 5]


class Std:
  def __init__(self, array: list[float]):
    self.array = array
    self.memuza = sum(array) / len(array)
    self.std = self.__get_std(array)
    self.most_from = self.memuza - self.std
    self.most_to = self.memuza + self.std

  def __get_std(self, array: list[float]) -> float:
    std = (sum(([(i-self.memuza)**2 for i in array]))/len(array))**0.5
    return std
  def __str__(self):
    return f"Most from {round(self.most_from, 2)} to {round(self.most_to, 2)}"


std_obj = Std(array)

print(std_obj)
```
