# Super nice gradient decent realization in python

```python
import math
from typing import Callable

# function of nigzeret
def nigzeret(x: float) -> float:
  high1 = (4*(x**3) + 15*(x**2) + 40 * x + 80) * (x**2 + 8 * x + 35)
  high2 = (2*x + 8) * (x**4 + 5 * (x ** 3) + 20 * (x ** 2) + 80 * x - 16)
  low = (x ** 2 + 8 * x + 35) ** 2

  return (high1 - high2) / low



def gradient_decent(nigzeret: Callable[[int],int], start: int, alpha: float, iter: int, leave: bool, time_log: int) -> float:
  # assighning unreal previous to not quit from the start
  previous = start + 999999
  # starting trying from start point
  x = start

  # entering cycle for iter
  for i in range(iter):
    # counting nigzerert volume from the given function
    value = nigzeret(x)

    # checking if we need to log somethink
    if i % (iter // time_log) == 0:
      print(f"value: {value}, x: {x}")

    # checking if we need to quit after different is smaller than alpha
    if math.fabs(previous - value) < alpha and leave:
      print(previous - value, alpha)
      return x
    # assighning new x value
    x -= alpha * value

    previous = value
  return x

# COUNTING MINIMUM POINT USING gradient_decent()
#minimum = gradient_decent(nigzeret=nigzeret, start=3, alpha=0.0001, iter=1000000, leave=False, time_log=5)
# LOGGING IT NICE
#print(f"Minimum i got: {round(minimum, 4)} and its nigzeret is: {round(nigzeret(minimum), 5)}")



# ===EXAMPLE WITH OTHER NIGZERET FUNCTION===

def easier_nigzeret(x) -> float:
  return 5 * (x ** 2) + 7 * x + 9 + 10 * (x ** 3)

def other_nigzeret(x: float) -> float:
    # Define a new complex expression as a mathematical function of x.
    # You can adjust this based on the function you want to minimize.

    numerator = (3 * x ** 4) + (10 * x ** 3) + (5 * x ** 2) - 20 * x + 50
    denominator = (x ** 2 + 5 * x + 6) ** 2 + 10

    return numerator / denominator


# COUNTING MINIMUM POINT USING gradient_decent()
minimum = gradient_decent(nigzeret=other_nigzeret, start=3, alpha=0.001, iter=10000000, leave=False, time_log=5)
# LOGGING IT NICE
print(f"Minimum i got: {round(minimum, 4)} and its nigzeret is: {round(other_nigzeret(minimum), 5)}")
```
