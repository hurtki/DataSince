```python

def maxl(array):
  max = -9999
  for i in array:
    if max < i:
      max = i
  return max


def minl(array):
  min = 9999
  for i in array:
    if min > i:
      min = i
  return min

def max_min_memuza(list1, list2, list3):
  return_list = [maxl(list1), minl(list2), sum(list3) / len(list3)]
  return return_list

print(max_min_memuza([1,2,3], [4,5,6], [7,8,9]))
```
