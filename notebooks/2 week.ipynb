```python
import math as m
# it's actually the num like 3**n = 9 => n = 2


def lg(num1, num2: float) -> float:
    return m.log(num1, num2)

```

```python
import seaborn as sbs
import matplotlib.pyplot as plt
import numpy as np

x = []
y = []

for i in np.arange(0.1, 1000.0, 0.01):
  X = i
  Y = lg(i, 2)
  x.append(X)
  y.append(Y)


sbs.scatterplot(x=x, y=y, size=0.3, color="r")
plt.show()
```
