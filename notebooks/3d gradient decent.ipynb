# derivative gradiet decent

```python
def f(x1,x2: float) -> float:
  return x1 ** 2 + 6 * x1 + x1 * x2 + x2 ** 2 + 12

def dx1(x1, x2: float) -> float:
  return 2 * x1 + 6 + x2

def dx2(x1, x2: float) -> float:
  return x1 + 2 * x2


x1 = 5
x2 = 5
alpha = 0.1

for i in range(50000):
  x1, x2 = x1 - alpha * dx1(x1, x2), x2 - alpha * dx2(x1, x2)


print(f'x1 = {round(x1, 4)}, x2 = {round(x2, 4)}, f(x1, x2) = {round(f(x1, x2), 4)}')
```
