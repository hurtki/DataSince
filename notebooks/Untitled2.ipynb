```python
import numpy as np
from scipy.optimize import minimize

# Define your nigzeret function
def nigzeret(x: float) -> float:
    high1 = (4*(x**3) + 15*(x**2) + 40 * x + 80) * (x**2 + 8 * x + 35)
    high2 = (2*x + 8) * (x**4 + 5 * (x ** 3) + 20 * (x ** 2) + 80 * x - 16)
    low = (x ** 2 + 8 * x + 35) ** 2

    return (high1 - high2) / low

# Use scipy's minimize function to find the minimum
result = minimize(nigzeret, x0=3)  # Starting at x=3

# Print the results
print(f"Minimum found at x = {result.x[0]:.4f} with function value {result.fun:.5f}")
```
